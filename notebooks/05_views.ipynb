```python
%sql
CREATE OR REPLACE VIEW streaming_project.ecommerce.vw_order_summary
AS
SELECT
    o.order_id,
    o.order_date,
    o.customer_id,
    o.product_id,
    o.quantity,
    p.title     AS product_name,
    p.category,
    p.price,
    ROUND(o.quantity * p.price, 2) AS line_total
FROM streaming_project.ecommerce.silver_orders o
LEFT JOIN streaming_project.ecommerce.silver_products p
    ON o.product_id = p.product_id;
```
